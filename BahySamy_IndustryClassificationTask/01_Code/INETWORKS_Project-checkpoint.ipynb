## extracting features and labels

```python
# extracting features and labels
features = df['job title']
labels = df['industry']
```

```python
# mapping diffrent values of labels
labels = labels.map({'IT':0, 'Marketing': 1, 'Education':2, 'Accountancy':3})
labels.value_counts()
```

### Bag of words concept 
(BoW) concept is a term used to specify the problems that have a 'bag of words' or a collection of text data that needs to be worked with. The basic idea of BoW is to take a piece of text and count the frequency of the words in that text. It is important to note that the BoW concept treats each word individually and the order in which the words occur does not matter.

```python
# Bagging words of job title columns
count_vector = CountVectorizer()
features = count_vector.fit_transform(features)
```

## Spliting Data

```python
X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size = .2, shuffle = True, stratify = labels)
print('length of X_train', X_train.shape)
print('length of X_test', X_test.shape)
print('length of y_train', y_train.shape)
print('length of y_test', y_test.shape)
```

## Creating Model & Fitting & Making Predictions
we will be using the multinomial Naive Bayes algorithm. This particular classifier is suitable for classification with discrete features (such as in our case, word counts for industry classification). It takes in integer word counts as its input.

```python
naive_bayes = MultinomialNB()
naive_bayes.fit(X_train, y_train)
```

```python
predictions = naive_bayes.predict(X_test)
```

```python
naive_bayes.score(X_train, y_train)
```

## Model Evaluation 

```python
print('Accuracy score: ', format(accuracy_score(y_test, predictions)))
print('Precision score: ', format(precision_score(y_test, predictions, average = None)))
print('Recall score: ', format(recall_score(y_test, predictions, average = None)))
print('F1 score: ', format(f1_score(y_test, predictions, average = None)))
```

## Saving & Reloading model and count_vector

```python
# saving model
dump(naive_bayes, 'model.joblib')
```

```python
# saving count vector
dump(count_vector, 'count_vector.joblib')
```

```python
# reload model
model_in = load('model.joblib')
model_in.score(X_train, y_train)
```
